```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the age ranges
age_ranges = ["18-25", "26-35", "36-45", "46-55", "55+"]

# Calculate the counts for each age group
age_counts = [
    len(df[(df['Age'] >= 18) & (df['Age'] <= 25)]),
    len(df[(df['Age'] >= 26) & (df['Age'] <= 35)]),
    len(df[(df['Age'] >= 36) & (df['Age'] <= 45)]),
    len(df[(df['Age'] >= 46) & (df['Age'] <= 55)]),
    len(df[df['Age'] > 55])
]

# Create the bar plot
plt.figure(figsize=(15, 6))
sns.barplot(x=age_ranges, y=age_counts, palette="mako")
plt.xlabel('Age Groups')
plt.ylabel('Count')
plt.title('Count of Customers in Each Age Group')
plt.show()
```

```python
# Define the spending score ranges
spending_score_ranges = ["1-20", "21-40", "41-60", "61-80", "81-100"]

# Calculate the counts for each spending score range
ss_counts = [
    len(df[(df["Spending Score (1-100)"] >= 1) & (df["Spending Score (1-100)"] <= 20)]),
    len(df[(df["Spending Score (1-100)"] >= 21) & (df["Spending Score (1-100)"] <= 40)]),
    len(df[(df["Spending Score (1-100)"] >= 41) & (df["Spending Score (1-100)"] <= 60)]),
    len(df[(df["Spending Score (1-100)"] >= 61) & (df["Spending Score (1-100)"] <= 80)]),
    len(df[(df["Spending Score (1-100)"] >= 81) & (df["Spending Score (1-100)"] <= 100)])
]

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=spending_score_ranges, y=ss_counts, palette="rocket")
plt.xlabel('Spending Score Ranges')
plt.ylabel('Count')
plt.title('Count of Customers in Each Spending Score Range')
plt.show()
```

```python
# Define the annual income ranges
income_ranges = ["$0 - 30,000", "$30,001-60,000", "$60,001-90,000", "$90,001-120,000", "$120,000-150,000"]

# Calculate the counts for each annual income range
income_counts = [
    len(df[(df["Annual Income (k$)"] >= 0) & (df["Annual Income (k$)"] <= 30)]),
    len(df[(df["Annual Income (k$)"] >= 31) & (df["Annual Income (k$)"] <= 60)]),
    len(df[(df["Annual Income (k$)"] >= 61) & (df["Annual Income (k$)"] <= 90)]),
    len(df[(df["Annual Income (k$)"] >= 91) & (df["Annual Income (k$)"] <= 120)]),
    len(df[(df["Annual Income (k$)"] >= 121) & (df["Annual Income (k$)"] <= 150)])
]

# Create the bar plot
plt.figure(figsize=(15, 6))
sns.barplot(x=income_ranges, y=income_counts, palette="Spectral")
plt.xlabel('Annual Income Ranges')
plt.ylabel('Count')
plt.title('Count of Customers in Each Annual Income Range')
plt.show()
```

```python
X1 = df.loc[:, ["Age", "Spending Score (1-100)"]].values
from sklearn.cluster import KMeans

wcss = []  # Initialize an empty list to store the Within-Cluster Sum of Squares (WCSS)

for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init="k-means++")  # Corrected the parameter names
    kmeans.fit(X1)
    wcss.append(kmeans.inertia_)

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))  # Corrected the function name and parameters
plt.grid()
plt.plot(range(1, 11), wcss, linewidth=2, color="red", marker="8")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("WCSS (Within-Cluster Sum of Squares)")
plt.title("Elbow Method for Optimal K")
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4)  # Initialize KMeans with the desired number of clusters (e.g., 4)
labels = kmeans.fit_predict(X1)  # Fit the KMeans model to the data and get cluster labels

print(labels)  # Print the cluster labels

```

```python
print(kmeans.cluster_centers_)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Create a KMeans instance with the desired number of clusters
kmeans = KMeans(n_clusters=4)

# Fit the KMeans model to the data and get cluster labels
labels = kmeans.fit_predict(X1)

# Get cluster centers (centroids)
cluster_centers = kmeans.cluster_centers_

# Create a scatter plot
plt.scatter(X1[:, 0], X1[:, 1], c=labels, cmap="rainbow", marker='o', s=50, alpha=0.7)
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1,], color="black", marker='X', s=200)

plt.title("Clusters of Customers")
plt.xlabel("Age")
plt.ylabel("Spending Score (1-100)")
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans

# Create a KMeans instance with the desired number of clusters (e.g., 5)
kmeans = KMeans(n_clusters=5)
clusters = kmeans.fit_predict(X3)

# Add the cluster labels to the DataFrame
df["label"] = clusters

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# Scatter plot for each cluster
colors = ["blue", "red", "green", "orange", "purple"]
for i in range(5):
    ax.scatter(
        df.Age[df.label == i],
        df["Annual Income (k$)"][df.label == i],
        df["Spending Score (1-100)"][df.label == i],
        c=colors[i],
        s=60,
        label=f'Cluster {i}'
    )

ax.set_xlabel("Age")
ax.set_ylabel("Annual Income (k$)")
ax.set_zlabel("Spending Score (1-100)")
plt.title("Clusters of Customers")
plt.legend()
plt.show()
```
